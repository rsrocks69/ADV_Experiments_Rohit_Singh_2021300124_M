### 1. Importing Libraries and Dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### 2. Data Preprocessing

```python
data = pd.read_csv(
    "../Datasets/birds.csv",
    delimiter="\t",
    nrows=None,
    usecols=[
        "species",
        "locality",
        "stateProvince",
        "individualCount",
        "decimalLatitude",
        "decimalLongitude",
        "day",
        "month",
        "year",
    ],
)
print(data.shape)
data.head()
```

```python
def set_dtypes(df):
    df["individualCount"] = df["individualCount"].astype("float32")
    df["decimalLatitude"] = df["decimalLatitude"].astype("float32")
    df["decimalLongitude"] = df["decimalLongitude"].astype("float32")
    df["day"] = df["day"].astype("int8")
    df["month"] = df["month"].astype("int8")
    df["year"] = df["year"].astype("int16")
    df["stateProvince"] = df["stateProvince"].astype("category")
    df["species"] = df["species"].astype("category")
    return df
```

```python
data = set_dtypes(data)
data.info()
```

```python
# remove rows where locality and count is missing
data = data.dropna(subset=["locality", "individualCount"])
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
# save the cleaned data
data.to_csv("../Datasets/birds_cleaned.csv", index=False)
```

```python
# sample the data
data = pd.read_csv("../Datasets/birds_cleaned.csv")
data = data.sample(frac=0.1, random_state=42)
data.info()
```

```python
# add a new column called "event_date" which is a combination of day, month and year
data["event_date"] = pd.to_datetime(data[["year", "month", "day"]])
data.head()
```

```python
# save the sampled data
data.to_csv("../Datasets/birds_sampled.csv", index=False)
```
