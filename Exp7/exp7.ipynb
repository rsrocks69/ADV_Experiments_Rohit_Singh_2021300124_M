### Hypothesis Testing

#### 1. Using T-test Statistic:
H0: There is no significant difference in annual incomes between Charged Off and Fully Paid loan statuses.  
HA: There is a significant difference in annual incomes between Charged Off and Fully Paid loan statuses.

```python
# print unique values of 'Loan Status'
print(data['Loan Status'].unique())
```

```python
# Filter data by 'Loan Status'
charged_off = data[data["Loan Status"] == "Charged Off"]["Annual Income"]
fully_paid = data[data["Loan Status"] == "Fully Paid"]["Annual Income"]
```

```python
# Perform the two-sample t-test
t_stat, p_value = stats.ttest_ind(charged_off, fully_paid)
```

```python
# Output the results of the t-test
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

```python
# Interpretation
if p_value < 0.05:
    print(
        "Reject the null hypothesis: There is a significant difference in annual incomes between Charged Off and Fully Paid loan statuses."
    )
else:
    print(
        "Fail to reject the null hypothesis: There is no significant difference in annual incomes between Charged Off and Fully Paid loan statuses."
    )
```

#### 2. Using Pearson Correlation Coefficient:
H0: There is no significant linear relationship between Annual Income and Credit Score.  
HA: There is a significant linear relationship between Annual Income and Credit Score.

```python
filtered_data = data[["Annual Income", "Credit Score"]].dropna()
```

```python
# Calculate the Pearson correlation coefficient and p-value
corr_coeff, p_value = stats.pearsonr(
    filtered_data["Annual Income"], filtered_data["Credit Score"]
)
```

```python
# Output the results of the Pearson correlation
print(f"Pearson Correlation Coefficient: {corr_coeff}")
print(f"P-value: {p_value}")
```

```python
# Interpretation
if p_value < 0.05:
    print(
        "Reject the null hypothesis: There is a significant linear relationship between Annual Income and Credit Score."
    )
else:
    print(
        "Fail to reject the null hypothesis: There is no significant linear relationship between Annual Income and Credit Score."
    )
```
